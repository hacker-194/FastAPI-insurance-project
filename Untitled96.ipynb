```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report , accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df
```

```python
df["occupation"].unique()
```

```python
df["bmi"] = df["weight"] / (df["height"]**2)
```

```python
df.sample(5)
```

```python
def age_group(age):
  if age < 25 :
    return "young"
  elif age < 45 :
    return "adult"
  elif age < 60 :
    return "senior"
  elif age >= 60 :
    return "old"
```

```python
df["age_group"]=df["age"].apply(age_group)
```

```python
def lifestyle_risk(row):
  if row["smoker"] and row ["bmi"] > 30:
    return "high"
  if row["smoker"] and row ["bmi"] >27 :
    return "medium"
  else:
    return "low"
```

```python
df["lifestyle_risk"]=df.apply(lifestyle_risk,axis=1)
```

```python
tier_1_cities = ["Mumbai", "Delhi", "Bangalore", "Chennai", "Kolkata", "Hyderabad", "Pune"]
tier_2_cities = [
    "Jaipur", "Chandigarh", "Indore", "Lucknow", "Patna", "Ranchi", "Visakhapatnam", "Coimbatore",
    "Bhopal", "Nagpur", "Vadodara", "Surat", "Rajkot", "Jodhpur", "Raipur", "Amritsar", "Varanasi",
    "Agra", "Dehradun", "Mysore", "Jabalpur", "Guwahati", "Thiruvananthapuram", "Ludhiana", "Nashik",
    "Allahabad", "Udaipur", "Aurangabad", "Hubli", "Belgaum", "Salem", "Vijayawada", "Tiruchirappalli",
    "Bhavnagar", "Gwalior", "Dhanbad", "Bareilly", "Aligarh", "Gaya", "Kozhikode", "Warangal",
    "Kolhapur", "Bilaspur", "Jalandhar", "Noida", "Guntur", "Asansol", "Siliguri"
]
```

```python
def city_tier(city):
  if city in tier_1_cities:
    return "tier_1"
  elif city in tier_2_cities:
    return "tier_2"
  else:
    return "error"
```

```python
df["city_tier"]=df["city"].apply(city_tier)
```

```python
df.sample(5)
```

```python
df.drop(columns=["age","weight","height","smoker"])
```

```python
df.isnull().sum()
```

```python
X = df[["bmi","age_group","lifestyle_risk","city_tier","income_lpa","occupation"]]
y = df["insurance_premium_category"]
```

```python
X
```

```python
y
```

```python
categorical_features = ["age_group","lifestyle_risk","city_tier","occupation"]
numerical_features = ["bmi","income_lpa"]
```

```python
preprocessor= ColumnTransformer(
    transformers=[
        ("cat",OneHotEncoder(),categorical_features),
        ("num","passthrough",numerical_features)
    ]
)
```

```python
pipeline = Pipeline(steps=[
    ("preprocessor",preprocessor),
    ("classifier",RandomForestClassifier(random_state=42))
])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
pipeline.fit(X_train,y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
